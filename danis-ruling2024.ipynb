## **Oakden: Yip vs. Bao**

**nb**: there is no ordering relation, so in the definitions below a e.g. HL contour is identical to a LH contour.

```python
yip_f = {
    'L' : {('s','-u'), ('-u','l')},
    'H' : {('s','+u'), ('+u','h')},
    'M1' : {('s','-u'), ('-u','h')},
    'M2' : {('s','+u'), ('+u','l')},
    'HM' : {('s','+u'), ('+u','h'), ('+u','l')},
    'MH' : {('s','+u'), ('+u','l'), ('+u','h')},
    'ML' : {('s','-u'), ('-u','l'), ('-u','h')},
    'LM' : {('s','-u'), ('-u','l'), ('-u','h')}
}

bao_f = {
    'L' : {('s','T'), ('T','-u'), ('T','c'), ('c','l')},
    'H' : {('s','T'), ('T','+u'), ('T','c'), ('c','h')},
    'M1' : {('s','T'), ('T','-u'), ('T','c'), ('c','h')},
    'M2' : {('s','T'), ('T','+u'), ('T','c'), ('c','l')},
    'HM' : {('s','T'), ('T','+u'), ('T','c'), ('c','h'), ('c','l')},
    'MH' : {('s','T'), ('T','+u'), ('T','c'), ('c','h'), ('c','l')},
    'ML' : {('s','T'), ('T','-u'), ('T','c'), ('c','h'), ('c','l')},
    'LM' : {('s','T'), ('T','-u'), ('T','c'), ('c','h'), ('c','l')},
}

yip = ax.Theory(yip_f, name="yip")
bao = ax.Theory(bao_f, name="bao")

oakden = ax.Comparison([yip, bao])
```
